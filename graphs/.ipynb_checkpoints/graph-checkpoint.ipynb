```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
log = pd.read_csv("../build/log.csv")
log.head()
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlabel("time step, k")
plt.ylabel("Cross track error")
plt.grid()
plt.plot(log["cte"], '-', color="c",label="cte")
plt.hlines(0,0,len(log["cte"]), color="r",label="reference cte")
plt.legend(loc=1) 
plt.title("Cross Track Error")
plt.show()
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlabel("time step, k")
plt.ylabel("Orientation error (radian)")
plt.grid()
plt.plot(log["epsi"], '-', color="c",label="epsi")
plt.hlines(0,0,len(log["epsi"]), color="r",label="reference epsi")
plt.legend(loc=1) 
plt.title("Orientation Error")
plt.show()
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlabel("time step, k")
plt.ylabel("Speed (mph)")
plt.grid()
plt.plot(log["v"], '-', color="c",label="speed")
plt.hlines(100,0,len(log["v"]), color="r",label="reference speed")
plt.legend(loc=5) 
plt.title("Speed")
plt.show()
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlabel("time step, k")
plt.ylabel("Steering angle (radian)")
plt.grid()
plt.plot(log["steer"], '-', color="c",label="steering angle")
plt.legend(loc=1) 
plt.title("Steering angle")
plt.show()
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlabel("time step, k")
plt.ylabel("Throttle")
plt.grid()
plt.plot(log["throttle"], '-', color="c",label="throttle")
plt.legend(loc=5) 
plt.title("Throttle")
plt.show()
```
